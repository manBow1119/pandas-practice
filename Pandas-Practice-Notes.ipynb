## Dealing with Incomplete Data

```python
#import data- notice the NaNs
car_sales_missing = pd.read_csv("car-sales-missing-data.csv")
car_sales_missing
```

```python
# fill nan values with the mean of the rest
car_sales_missing["Odometer"].fillna(car_sales_missing["Odometer"].mean(), inplace=True)

#broken into steps with stored value-dropna not necessary
# odom_mean = car_sales_missing["Odometer"].dropna().mean()
# car_sales_missing["Odometer"].fillna(odom_mean)

car_sales_missing
```

```python
# remove all other missing data
car_sales_missing.dropna(inplace=True)
car_sales_missing
```

```python
#extra copy to build on
car_sales_trim = car_sales_missing
car_sales_trim
```

```python
# Add seats column and assign all to default of 5
car_sales_trim["Seats"] = 5
car_sales_trim
```

```python
#import and test random
import random
random.random() + random.randrange(1,4)-.5
```

```python
 # create engine size column and give random engine sizes between 1.3 and 4.5
car_sales_trim["Engine Size"] = [(random.random() + random.randrange(2,5)-.7), (random.random() + random.randrange(2,5)-.7), (random.random() + random.randrange(2,5)-.7), (random.random() + random.randrange(2,5)-.7), (random.random() + random.randrange(2,5)-.7), (random.random() + random.randrange(2,5)-.7)]
car_sales_trim
```

```python
# create new column for $/km- must reformat price for new df
car_sales_trim["Price"] = car_sales_trim["Price"].str.replace("[\$\,]|\.\d*", '').astype(int)

car_sales_trim["Price per km"] = car_sales_trim["Price"]/car_sales_trim["Odometer"]
car_sales_trim
```

```python
#remove column, axis, inplace
car_sales_trim.drop(["Price per km"], axis=1, inplace=True)
car_sales_trim
```

```python
# use sample and manipulate the 'frac' arg to shuffle or randomnly select
car_sales_trim_shuf = car_sales_trim.sample(frac=1)
car_sales_trim_shuf
```

```python
#Reset indices
car_sales_trim_shuf.reset_index(drop=True)
```

```python
#use lambda func to convert odometer to miles
car_sales_trim_shuf["Odometer"] = car_sales_trim_shuf["Odometer"].apply(lambda x: x/1.6).round()
car_sales_trim_shuf["Engine Size"] = car_sales_trim_shuf["Engine Size"].round(1)
car_sales_trim_shuf
```
